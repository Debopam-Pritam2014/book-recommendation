# Popularity Based Recommender System

top 50 books having highest avg ratings, minimum 250 ratings

```python
rating_with_bookName=ratings.merge(books,on='ISBN')
```

```python
rating_with_bookName.head()
```

```python
rating_with_bookName.info()
```

```python
rating_with_bookName.isnull().sum()
```

```python
num_ratings_with_bookName=rating_with_bookName.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings_with_bookName.rename(columns={'Book-Rating':'Num-Ratings'},inplace=True)
num_ratings_with_bookName
```

```python
rating_with_bookName['Book-Rating']=rating_with_bookName['Book-Rating'].astype('int')
```

```python
avg_ratings_with_bookName=rating_with_bookName.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_ratings_with_bookName.rename(columns={'Book-Rating':'Avg-Ratings'},inplace=True)
avg_ratings_with_bookName
```

```python
avg_ratings_with_bookName[avg_ratings_with_bookName['Avg-Ratings']>=10]
```

```python
popular_df=num_ratings_with_bookName.merge(avg_ratings_with_bookName,on='Book-Title')
popular_df
```

```python
most_popular_df=popular_df[popular_df['Num-Ratings']>=200].sort_values(by='Avg-Ratings',ascending=False).head(50)
```

```python
final_popular_df=most_popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Num-Ratings','Avg-Ratings','Year-Of-Publication','Image-URL-M']]
```

```python
final_popular_df=final_popular_df.reset_index(drop=True)
final_popular_df
```

# Collaborative Filtering Based Recommender System

users give more than 200 books reviews, books that gain atleast 50 ratings from users

```python
rating_with_bookName.head(2)
```

```python
x=rating_with_bookName.groupby('User-ID').count()['Book-Rating']>=200
x=x[x]
```

```python
genuine_users=x.index
```

```python
filtered_ratings=rating_with_bookName[rating_with_bookName['User-ID'].isin(genuine_users)]
```

```python
y=filtered_ratings.groupby('Book-Title').count()['Book-Rating']>50
famous_books=y[y].index
```

```python
final_ratings_df=filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
pt=final_ratings_df.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Simialrity_matrix=cosine_similarity(pt)
```

```python
Simialrity_matrix.shape
```

```python
Simialrity_matrix
```

```python
def recommend(book_name):
    # fetch index
    book_index=np.where(pt.index==book_name)[0][0]
    similar_books_index=sorted(list(enumerate(Simialrity_matrix[book_index,:])),key=lambda x:x[1],reverse=True)[1:6]
    suggested_books=[]
    for book in similar_books_index:
        suggested_books.append(pt.index[book[0]])
        temp_df=books[books['Book-Title']==pt.index[book[0]]]
        print(temp_df.drop_duplicates('Book-Title')['Book-Author'])
    return suggested_books
    
```
